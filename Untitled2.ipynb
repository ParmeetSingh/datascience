```python
import numpy as np
from nltk.stem import WordNetLemmatizer
import random
import theano 
import theano.tensor as T
import brown
from bs4 import BeautifulSoup as bs
import nltk
```

```python
positive_reviews = bs(open('datasets/positive.review').read())
positive_reviews = positive_reviews.findAll('review_text')
```

```python
stopwords = set([word.strip() for word in open('datasets/stopwords.txt')])
stopwords = stopwords.union({
    'introduction', 'edition', 'series', 'application',
    'approach', 'card', 'access', 'package', 'plus', 'etext',
    'brief', 'vol', 'fundamental', 'guide', 'essential', 'printed',
'third', 'second', 'fourth', })
def myCustomTokenizer(line):
    line = line.lower()
    words = nltk.tokenize.word_tokenize(line)
    words = [word for word in words if word not in stopwords]
    lemmatizer = WordNetLemmatizer()
    words = [lemmatizer.lemmatize(word) for word in words]
    words = [word for word in words if len(word)>2]
    words = [word for word in words if not any(c.isdigit() for c in word)]
    return words
word2idx = {'START':0,'END':1}
current = 0
sentences = []
for review in positive_reviews:
    review = review.text.lower()
    tokens = myCustomTokenizer(review)
    for token in tokens:
        if token not in word2idx:
            word2idx[token] = current
            current += 1
        sentences.append(word2idx[token])
```

```python
V = len(word2idx)
N = len(sentences)
D = 100
```

```python
class rnn():
    def __init__(self,M,D):
        #dimension of word embedding layer
        self.D = D
        #dimension of the hidden layer
        self.M = M
        #dimension of vocabulary
        self.V = V
    def init_weight(self,Mi,Mo):
        return np.random.randn(Mi,Mo)
    def fit(self,X,alpha=10e-4,print_period=10,mu=0.99,activation=T.nnet.relu,batchsize = 10):
        
        #initializing all weights
        V = self.V
        D = self.D
        M = self.M
        self.f = activation
        
        We = theano.shared(self.init_weight(V,D))
        Wh = theano.shared(self.init_weight(D,M))
        bh = theano.shared(np.zeros(M))
        Wo = theano.shared(self.init_weight(M,V))
        bo = theano.shared(np.zeros(V))
        
        thX = T.ivector('thX')
        thY = T.ivector('thY')
        thE = We[thX] 
        
        #forward pass
        h = self.f(T.dot(thE,Wh) + bh)
        y = T.nnet.softmax(T.dot(h,Wo) + bo)
        
        prediction = T.argmax(y,axis=1)
        
        cost = -1*T.mean(T.log(y[T.arange(thY.shape[0]),thY])) 
    
        self.params = [We,Wh,bh,Wo,bo]
        
        self.dparams  = [theano.shared(p.get_value()*0) for p in self.params]
        
        self.grad = T.grad(cost,self.params)
        
        updates = [(p,p + mu*dp - alpha*g) for p,dp,g in zip(self.params,self.dparams,self.grad)] + [
            (dp,mu*dp - alpha*g) for dp,g in zip(self.dparams,self.grad)
        ]
        
        self.train = theano.function(
            inputs = [thX,thY],
            outputs = [prediction,cost],
            updates = updates
        )
        
        iterations = len(X)/batchsize
        
        Y = X + [1]
        X = [0] + X
        print len(X)
        print len(Y)
        num_batches = 2
        for i in xrange(num_batches):
            for x in xrange(iterations):
                xbatch = X[x*batchsize:x*batchsize + batchsize]
                ybatch = Y[x*batchsize:x*batchsize + batchsize]
                print "Start"
                prediction,cost = self.train(xbatch,ybatch)
                print "Cost is",cost
```

```python
rn = rnn(5,D)
rn.fit(sentences)
```

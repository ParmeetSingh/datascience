```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = np.random.randn(100,2) + [2,2]
data2 = np.random.randn(100,2) + [-2,2]
data3 = np.random.randn(100,2) + [0,0]
```

```python
plt.scatter(data[:,0],data[:,1],marker='*',c='red',alpha=0.5)
plt.scatter(data2[:,0],data2[:,1],marker='*',c='blue',alpha=0.5)
plt.scatter(data3[:,0],data3[:,1],marker='*',c='green',alpha=0.5)
```

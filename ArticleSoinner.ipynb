```python
import numpy as np
from bs4 import BeautifulSoup as bs
import nltk
import random
```

```python
positive_reviews = bs(open('datasets/positive.review').read())
positive_reviews = positive_reviews.findAll('review_text')
```

```python
trigrams = {}
for review in positive_reviews:
    review = review.text.lower()
    tokens = nltk.tokenize.word_tokenize(review)
    length = len(tokens)
    for i in xrange(length-2):
        if (tokens[i],tokens[i+2]) not in trigrams:
            trigrams[(tokens[i],tokens[i+2])] = []
        trigrams[(tokens[i],tokens[i+2])].append(tokens[i+1])
```

```python
trigrams_word_count={}
for key,token_list in trigrams.iteritems():
    #dictionary creation
    d = {}
    total_words = 0
    if len(set(token_list)) > 1:
        for word in token_list:
            if word not in d:
                d[word]=0
            d[word] += 1
            total_words += 1
        #division by total to make probabalities
        for key2,value in d.iteritems():
            d[key2] /= 1.0*total_words
        trigrams_word_count[key]=d
        print key
```
